```python
import random
x = 1
sum = 0
thrownDiceNumbers = []


while x < 6:
    thrownDiceNumbers.append(random.randint(1,6))
    x = x + 1

print (thrownDiceNumbers)

for i in thrownDiceNumbers:
    sum = sum + i
print ("Sum of the rolls is: " + str(sum))

iso = max(thrownDiceNumbers)
print (iso)
```
